```python
# Imports the Pandas library and stores the dataset into a dataframe
import pandas as pd
poll_url = 'http://projects.fivethirtyeight.com/general-model/president_general_polls_2016.csv'
polls = pd.read_csv(poll_url)
```

```python
# Displays the first 5 rows of the data
polls.head()
```

```python
# Sorts the dataframe according to column name 'state' and displays first 2 rows
polls.sort_values('state', inplace=True)
polls.head(2)
```

## Clean the data

```python
polls.sort_values('state', ascending=False).head
```

### Keep only the rows with "now-cast" in the type column

```python
polls = polls.query('type=="now-cast"')
polls.head()
```
