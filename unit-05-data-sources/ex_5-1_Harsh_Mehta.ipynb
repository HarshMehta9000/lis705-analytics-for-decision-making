```python
# Imports the Pandas library and stores the dataset into a dataframe
import pandas as pd
```

```python
#The URL of the dataset containing polling data for the 2016 U.S. Presidential Election.
polls_url = (
    'http://projects.fivethirtyeight.com/general-model/president_general_polls_2016.csv')
```

```python
# Import the request module from urllib, a package that allows for opening and reading URLs
from urllib import request
# Use the urlretrieve function from the request module to download the dataset from the specified URL.
# The dataset is saved locally with the filename 'president_polls_2016.csv'.
# This process involves specifying the source URL and the desired local filename for the saved file.
request.urlretrieve(polls_url, filename ='president_polls_2016.csv')
```

```python
# Load the dataset containing the 2016 U.S. Presidential Election polls from a CSV file into a pandas Polls DataFrame.
polls = pd.read_csv('president_polls_2016.csv')
```
