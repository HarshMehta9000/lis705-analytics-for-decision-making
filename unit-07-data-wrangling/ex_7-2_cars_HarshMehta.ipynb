## Read the data

```python
# Import the pandas library
import pandas as pd
```

```python
# Load the dataset from a CSV file named 'cars.csv' into a DataFrame named 'cars'. This file contains information about various cars.
cars = pd.read_csv('cars.csv')
```

```python
# Display the first 5 rows of the 'cars' DataFrame to provide an overview of the data structure and contents.
cars.head()
```

## Add and drop columns

```python
# Retrieve and display the unique values in the 'CarName' column, which may include both brand and model information.
cars.CarName.unique()
```

```python
# Create a new column 'brand' by applying a lambda function to each row. This function splits the 'CarName' string on spaces and takes the first element, assumed to be the car's brand.
cars['brand'] = cars.apply(lambda x: x.CarName.split(' ')[0], axis=1)
```

```python
# Display the unique values in the 'brand' column to verify the extraction of car brands.
cars.brand
```

```python
# Define a function 'get_model' to extract the model name from the 'CarName' column. This function splits 'CarName' into parts, discards the first part (the brand), and joins the rest back into a string.
def get_model(row):
    names = row.CarName.split(' ')  # split into a list of names
    name = ' '.join(names[1:])      # join all names except the first, which is the brand
    return name
    # Apply the 'get_model' function to each row to create a new 'model' column in the 'cars' DataFrame.
cars['model'] = cars.apply(get_model, axis=1)

# Display the unique values in the 'model' column to verify the extraction of car models.
cars.model
```

```python
# Remove the columns 'car_ID' and 'CarName' from the DataFrame, as they're no longer needed. The 'inplace=True' parameter modifies the DataFrame in place.
cars.drop(columns=['car_ID','CarName'], inplace=True)
```

```python
# Display the first 5 rows of the modified 'cars' DataFrame to see the results of the column removals.
cars.head()
```

## Set an index and unstack the data

```python
# Create a new DataFrame 'cars_filtered' by filtering 'cars' for rows where the 'brand' is 'volkswagen' and selecting only the 'model', 'horsepower', 'carbody', and 'doornumber' columns.
cars_filtered = cars.query('brand == "volkswagen"')[['model','horsepower','carbody','doornumber']]

# Display the first 5 rows of 'cars_filtered' to inspect the filtered and selected data.
cars_filtered.head()
```

```python
# Set a MultiIndex for 'cars_filtered' using the 'model', 'carbody', and 'doornumber' columns. This organizes the data hierarchically by these attributes.
cars_indexed = cars_filtered.set_index(['model','carbody','doornumber'])

# Display the 'cars_indexed' DataFrame to see the structure with the newly set MultiIndex.
cars_indexed
```

```python
# Unstack the 'cars_indexed' DataFrame by the 'carbody' level of the index, rearranging the data so that 'carbody' values become column headings. This operation provides a different view of the data, making it easier to compare across 'carbody' types.
cars_indexed.unstack('carbody')
```
