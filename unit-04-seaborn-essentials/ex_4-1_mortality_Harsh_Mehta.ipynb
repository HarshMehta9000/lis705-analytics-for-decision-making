## The basic Seaborn parameters

```python
# Import Seaborn library for data visualization
import seaborn as sns
```

```python
# Created a line plot using Seaborn to visualize the mortality data with X axis as Year, Y axis as Death Rate and hue as Age Group
sns.relplot(data=mortality_data, kind='line',
            x='Year', y='DeathRate', hue='AgeGroup')
```

```python
#Display a relational plot with Seaborn, adjusting plot size and aspect ratio for better readability
sns.relplot(data=mortality_data, 
            kind='line', x='Year', y='DeathRate', hue='AgeGroup', 
            height=4, aspect=1.25, 
            col='AgeGroup', col_wrap=2, legend=False)
```

## Enhancing plots

```python
# Customized a line plot with Seaborn, including setting the axis labels and plot title for better readability
ax = sns.lineplot(data=mortality_data, 
                  x='Year', y='DeathRate', hue='AgeGroup')
ax.set(title='Deaths by Age Group', ylabel='Deaths per 100,000')
```

```python
# Created a relational plot with Seaborn, grouping data by 'AgeGroup' and customizing the figure size
g = sns.relplot(data=mortality_data, kind='line',
    x='Year', y='DeathRate', hue='AgeGroup', aspect=1.5)
for ax in g.axes.flat:
    ax.set(title='Deaths by Age Group', ylabel='Deaths per 100,000')
```

```python
# Changed cell 10 so every fifth year is displayed on the x-axis for the plot and so the labels on the x-axis are displayed at their default rotation.
g = sns.relplot(data=mortality_data, kind='line',
    x='Year', y='DeathRate', hue='AgeGroup', aspect=1.25)
for ax in g.axes.flat:
    ax.set(title='Deaths by Age Group (1910-1930)', ylabel='Deaths per 100,000', 
        xticks=[x for x in range(1910, 1931, 5)], 
        xlim=(1910,1930), ylim=(0,1750)),
    ax.tick_params('x', labelrotation=30)
```

```python
#Another example of a relational plot with Seaborn, this time focusing on a specific time range and adding plot titles
ax = sns.lineplot(data=mortality_data,
     x='Year', y='DeathRate', hue='AgeGroup')
ax.set(title='Deaths by Age Group (1910-1930)', 
      ylabel='Deaths per 100,000', 
      xticks=[x for x in range(1910,1931,2)], 
      xlim=(1910,1930), ylim=(0,1750))
ax.tick_params('x', labelrotation=30)
```

```python
# Demonstrating the use of lineplot with Seaborn to plot mortality data, customizing the legend
sns.set_style('whitegrid')
sns.relplot(data=mortality_data.query('Year >= 1910 and Year <= 1930'),
    kind='line', x='Year', y='DeathRate', hue='AgeGroup', aspect=1.25)
```

```python
#Create a relational plot with Seaborn, grouping data by 'AgeGroup' and customizing the figure size
sns.set_style('darkgrid')
sns.relplot(data=mortality_data.query('Year >= 1910 and Year <= 1930'),
    kind='line', x='Year', y='DeathRate', hue='AgeGroup', aspect=1.25)
```

```python
# Opt for a minimalistic look with 'ticks' style, showing ticks on the x and y axes
sns.set_style('ticks')
```

```python
# Created a relational plot for a specific time range (1910 to 1930), focusing on yearly mortality rates
g = sns.relplot(data=mortality_data.query('Year >= 1910 and Year <= 1930'),
    kind='line', x='Year', y='DeathRate', hue='AgeGroup', legend=False,
    col='AgeGroup', col_wrap=2, height=4)
g.fig.suptitle('Deaths by Age Group (1910-1930)', y=1.025)
g.set(ylabel='Deaths per 100,000', 
      xticks=[x for x in range(1910,1931,5)], 
      ylim=(0,1750))
```

## Relational plots

```python
#Plot mortality data focusing on years 1910 to 1930, showcasing how to filter data within Seaborn plots.
# We use the 'query' method to select a subset of the data and 'kind=line' to create line plots.
sns.relplot(data=mortality_data, kind='line', x='Year', y='DeathRate', 
            hue='AgeGroup')
```

```python
# Use Seaborn to create a line plot showing mortality rates over time, customizing the confidence interval representation
sns.relplot(data=mortality_data, kind='line', x='Year', y='DeathRate', ci='sd')
```

```python
# Display a scatter plot to examine the relationship between year and death rate, with different age groups represented by colors.
sns.relplot(data=mortality_data, kind='scatter', x='Year', y='DeathRate', 
            hue='AgeGroup')
```

```python
# Focus on a specific time range (1910-1930) with a scatter plot, comparing death rates across age groups
g = sns.relplot(data=mortality_data.query('Year >= 1910 and Year <= 1930'), 
    kind='scatter', x='Year', y='DeathRate', hue='AgeGroup', 
    size='DeathRate', sizes=(10,100))
for ax in g.axes.flat:
    ax.set(xticks=[x for x in range(1910,1931,2)])
```

## Categorical plots

```python
 #Changed cell 20 so it uses a specific method to create the bar plot, and change the confidence interval to 75.
sns.catplot(data=mortality_data.query('Year in (1950,2000)'), 
            kind='bar', x='Year', y='DeathRate', ci=75)
```

```python
# Create a horizontal bar plot using Seaborn's catplot to visualize the distribution of death rates for different age groups in the years 1950 and 2000.
# The 'query' function filters the dataset for these years, and 'orient='h'' makes the bars horizontal. 
# 'x' is set to 'DeathRate' to define the length of the bars, 'y' to 'Year' to categorize bars by year, and 'hue' to 'AgeGroup' for color coding different age groups.
sns.catplot(data=mortality_data.query('Year in (1950,2000)'), 
            kind='bar', x='DeathRate', y='Year', hue='AgeGroup', orient='h')
```

```python
# Generate a box plot using Seaborn's catplot to examine the distribution of death rates across the years 1915 to 1920.
# The 'query' method is used to filter the dataset to include only the years between 1915 and 1920.
# The plot categorizes data by 'Year' on the x-axis and shows the distribution of 'DeathRate' on the y-axis.
# Box plots are useful for visualizing the central tendency, dispersion, and outliers within the data.
sns.catplot(data=mortality_data.query('Year >= 1915 and Year <= 1920'), 
            kind='box', x='Year', y='DeathRate')
```

```python
# Generate a box plot using Seaborn's catplot to examine the distribution of death rates across the years 1915 to 1920.
# The 'query' method is used to filter the dataset to include only the years between 1915 and 1920.
# The plot categorizes data by 'Year' on the x-axis and shows the distribution of 'DeathRate' on the y-axis.
# Box plots are useful for visualizing the central tendency, dispersion, and outliers within the data.
sns.catplot(data=mortality_data,
            kind='box', x='DeathRate', y='AgeGroup', orient='h')
```

## Distribution plots

```python
# Generate a histogram using Seaborn's displot to visualize the overall distribution of death rates within the mortality data.
# The 'kind' parameter is set to 'hist' to specify that a histogram should be created, with 'x' set to 'DeathRate' to define the data being plotted.
# This plot provides insight into the frequency distribution of death rates, allowing for the identification of common ranges of death rates and any potential skewness in the data.
sns.displot(data=mortality_data, kind='hist', x='DeathRate')
```

```python
#Change cell 25 so the histogram uses twice as many bins as it currently does from 8 to double 16
sns.displot(data=mortality_data, kind='hist', x='DeathRate', bins=16)
```

```python
# Create a Kernel Density Estimate (KDE) plot with Seaborn to visualize the probability density of death rates in the mortality data.
sns.displot(data=mortality_data, kind='kde', x='DeathRate')
```

```python
# Generate an Empirical Cumulative Distribution Function (ECDF) plot to visualize the cumulative distribution of death rates, segmented by age group.
sns.displot(data=mortality_data, kind='ecdf', x='DeathRate', hue='AgeGroup')
```

```python
# Create a histogram with an overlayed Kernel Density Estimate (KDE) to analyze the distribution of death rates, using 8 bins for granularity.
sns.displot(data=mortality_data, kind='hist', 
            x='DeathRate', kde=True, bins=8)
```

```python
# Generate filled KDE plots for death rates by age group, organized into a 2-column layout without legends and with specified plot height.
sns.displot(data=mortality_data, kind='kde', x='DeathRate', hue='AgeGroup', 
            fill=True, col='AgeGroup', col_wrap=2, height=3, legend=False)
```

## Other techniques for enhancing a plot

```python
# Create a line plot to show death rates by age group from 2000 to 2016, customizing the plot appearance and axis labels for clarity.
g = sns.relplot(
    data=mortality_data, 
    kind='line', x='Year', y='DeathRate', hue='AgeGroup', aspect=1.25)
for ax in g.axes.flat:
    ax.set_title('Deaths by Age Group (2000-2016)')
    ax.set_xlabel('')
    ax.set_xticks([x for x in range(2000,2017,2)])
    ax.set_xticklabels(['Year ' + str(x) for x in range(0,17,2)])
    ax.tick_params('x', labelrotation=30)
    ax.set_ylabel('Deaths per 100,000')
    ax.set_xlim(2000,2016)
    ax.set_ylim(0,80)
```

```python
# Plot death rates by age group over years with annotations highlighting the Spanish Flu Pandemic peak in 1918.
ax = sns.lineplot(data=mortality_data, 
                  x='Year', y='DeathRate', hue='AgeGroup')
ax.set_title('Deaths by Age Group')
ax.set_ylabel('Deaths per 100,000')
ax.annotate(text='Spanish Flu Pandemic', 
    xy=(1918, 1650), xytext=(1925, 1900), 
    arrowprops=dict(facecolor='red', width=3, headwidth=12, headlength=6))
```

```python
# Set a custom color palette and create a line plot to visualize death rates by age group, using specified colors for clarity.
colors = ['black','red','orange','blue']
sns.set_palette(sns.color_palette(colors))
sns.relplot(data=mortality_data, kind='line',
            x='Year', y='DeathRate', hue='AgeGroup')
```

```python
# Reset the color palette to 'tab10' for subsequent plots, offering a diverse and visually distinct set of colors.
sns.set_palette(sns.color_palette('tab10'))
```

```python
# Create segmented line plots for death rates by age group between 1910 and 1930, customizing layout and axis details for enhanced readability.
g = sns.relplot(
    data=mortality_data.query('Year >= 1910 and Year <= 1930'),
    kind='line', x='Year', y='DeathRate', hue='AgeGroup', legend=False,
    col='AgeGroup', col_wrap=2, height=3, aspect=1.2)
g.fig.suptitle('Deaths by Age Group (1910-1930)', y=1.025)
for ax in g.axes.flat:
    ax.set_ylabel('Deaths per 100,000')
    ax.set_xticks([x for x in range(1910, 1931, 2)])
    ax.tick_params('x', labelrotation=45)   
```

```python
# Generate line plots for each age group's death rates from 1910 to 1930, customizing each subplot with titles and adjusting axis labels for clear interpretation.
g = sns.relplot(
    data=mortality_data.query('Year >= 1910 and Year <= 1930'), 
    kind='line', x='Year', y='DeathRate', hue='AgeGroup', legend=False, 
    col='AgeGroup', col_wrap=2, height=3, aspect=1.2)
g.fig.suptitle('Deaths by Age Group (1910-1930)', y=1.025)
ageGroups = mortality_data['AgeGroup'].drop_duplicates().tolist()
for index, ax in enumerate(g.axes.flat):
    ax.set_title(ageGroups[index])
    ax.set_ylabel('Deaths per 100,000')
    ax.set_xticks([x for x in range(1910, 1931, 2)])
    ax.tick_params('x', labelrotation=45)
```
